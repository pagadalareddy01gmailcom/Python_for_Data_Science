```python
import pandas as pd

df = pd.read_excel("/PAGADALA Tulasiram Reddy.xlsx")
```

```python

print("First 5 rows of the DataFrame:")
display(df.head())


print("\nDataFrame Info:")
df.info()


print("\nDescriptive Statistics:")
display(df.describe())


print("\nExploring Unique Values in Object Columns:")
for col in df.columns:
    if df[col].dtype == 'object':
        print(f"\nColumn: {col}")
        print(f"Number of unique values: {df[col].nunique()}")
        print(f"Unique values in '{col}':")
        print(df[col].unique())
```

```python
# Examine unique values in 'Domain' column for inconsistencies
print("Unique values in 'Domain' column before processing:")
print(df['Domain'].unique())

# Standardize 'tech:it' to 'Technology' for consistency
df['Domain'] = df['Domain'].replace('tech:it', 'Technology')

# Check unique values after standardization
print("\nUnique values in 'Domain' column after processing:")
print(df['Domain'].unique())

# Re-check data types and missing values to ensure no issues were introduced or missed
print("\nDataFrame Info after processing:")
df.info()

# Re-check descriptive statistics for numerical columns
print("\nDescriptive Statistics after processing:")
display(df.describe())

# Although outliers were not explicitly found in EDA,
# for numerical columns like 'Market value as of 31 December, 2023' and 'Number of shares'
# given the wide range and potential for a few large companies,
# it's reasonable to assume potential outliers.
# However, without specific instructions or context on how to handle them (e.g., capping, removal),
# and given the small dataset size, we will note their presence but not apply specific outlier treatment
# at this stage, as it might remove significant data points.
# This decision is based on the trade-off between removing potential noise and losing valuable information.
print("\nPotential outliers might exist in numerical columns with large ranges, but no specific handling is applied at this stage.")
```

```python
# 1. Calculate the total market value of all companies in the portfolio.
total_market_value = df['Market value as of 31 December, 2023'].sum()
print(f"Total Market Value of the Portfolio: ${total_market_value:,.2f}")

# 2. Determine the number of companies in each domain.
companies_by_domain = df['Domain'].value_counts()
print("\nNumber of Companies in Each Domain:")
display(companies_by_domain)

# 3. Calculate the average score for companies in each domain.
average_score_by_domain = df.groupby('Domain')['Score'].mean()
print("\nAverage Score for Companies in Each Domain:")
display(average_score_by_domain)

# 4. Find the top 5 companies by market value.
top_5_by_market_value = df.nlargest(5, 'Market value as of 31 December, 2023')
print("\nTop 5 Companies by Market Value:")
display(top_5_by_market_value[['Company', 'Market value as of 31 December, 2023']])

# 5. Find the bottom 5 companies by market value.
bottom_5_by_market_value = df.nsmallest(5, 'Market value as of 31 December, 2023')
print("\nBottom 5 Companies by Market Value:")
display(bottom_5_by_market_value[['Company', 'Market value as of 31 December, 2023']])

# 6. Calculate the correlation matrix for numerical columns.
# Select only numerical columns for correlation calculation
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
correlation_matrix = df[numerical_cols].corr()
print("\nCorrelation Matrix for Numerical Columns:")
display(correlation_matrix)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Bar plot for the number of companies in each domain
plt.figure(figsize=(10, 6))
sns.barplot(x=companies_by_domain.index, y=companies_by_domain.values)
plt.title('Number of Companies in Each Domain')
plt.xlabel('Domain')
plt.ylabel('Number of Companies')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 3. Bar plot for the average score for companies in each domain
plt.figure(figsize=(10, 6))
sns.barplot(x=average_score_by_domain.index, y=average_score_by_domain.values)
plt.title('Average Score by Domain')
plt.xlabel('Domain')
plt.ylabel('Average Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 4. Scatter plot to visualize the relationship between 'Market value as of 31 December, 2023' and 'Score'.
# Using a logarithmic scale for market value due to large range
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Market value as of 31 December, 2023', y='Score')
plt.xscale('log') # Use logarithmic scale for better visualization of wide range
plt.title('Market Value vs. Score (Log Scale for Market Value)')
plt.xlabel('Market value as of 31 December, 2023 (Log Scale)')
plt.ylabel('Score')
plt.tight_layout()
plt.show()

# 5. Histogram of the 'Score' distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Score', bins=10, kde=True)
plt.title('Distribution of Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# 6. Heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Variables')
plt.tight_layout()
plt.show()
```

```python
# Examine unique values in 'Domain' column for inconsistencies before processing.
print("Unique values in 'Domain' column before processing:")
print(df['Domain'].unique())

# Standardize the value 'tech:it' to 'Technology' for consistency in the 'Domain' column.
# This ensures that all technology-related companies are grouped under a single category.
df['Domain'] = df['Domain'].replace('tech:it', 'Technology')

# Check unique values after standardization to confirm the change was applied correctly.
print("\nUnique values in 'Domain' column after processing:")
print(df['Domain'].unique())

# Re-check data types and missing values after processing to ensure no issues were introduced or missed during cleaning.
print("\nDataFrame Info after processing:")
df.info()

# Re-check descriptive statistics for numerical columns.
# This is done to see if any processing steps (though none were applied to numerical columns here)
# might have inadvertently affected the numerical data distribution.
print("\nDescriptive Statistics after processing:")
display(df.describe())

# Although outliers were not explicitly found in EDA,
# for numerical columns like 'Market value as of 31 December, 2023' and 'Number of shares'
# given the wide range and potential for a few large companies,
# it's reasonable to assume potential outliers.
# However, without specific instructions or context on how to handle them (e.g., capping, removal),
# and given the small dataset size, we will note their presence but not apply specific outlier treatment
# at this stage, as it might remove significant data points.
# This decision is based on the trade-off between removing potential noise and losing valuable information.
print("\nPotential outliers might exist in numerical columns with large ranges, but no specific handling is applied at this stage.")
```

```python
# 1. Calculate the total market value of all companies in the portfolio by summing the 'Market value as of 31 December, 2023' column.
total_market_value = df['Market value as of 31 December, 2023'].sum()
# Print the calculated total market value, formatted for readability.
print(f"Total Market Value of the Portfolio: ${total_market_value:,.2f}")

# 2. Determine the number of companies in each domain by counting the occurrences of each unique value in the 'Domain' column.
companies_by_domain = df['Domain'].value_counts()
print("\nNumber of Companies in Each Domain:")
# Display the resulting Series.
display(companies_by_domain)

# 3. Calculate the average score for companies in each domain by grouping the DataFrame by 'Domain' and calculating the mean of the 'Score' column for each group.
average_score_by_domain = df.groupby('Domain')['Score'].mean()
print("\nAverage Score for Companies in Each Domain:")
# Display the resulting Series.
display(average_score_by_domain)

# 4. Find the top 5 companies by market value using the nlargest() method on the 'Market value as of 31 December, 2023' column.
top_5_by_market_value = df.nlargest(5, 'Market value as of 31 December, 2023')
print("\nTop 5 Companies by Market Value:")
# Display the 'Company' and 'Market value as of 31 December, 2023' columns for the top 5 companies.
display(top_5_by_market_value[['Company', 'Market value as of 31 December, 2023']])

# 5. Find the bottom 5 companies by market value using the nsmallest() method on the 'Market value as of 31 December, 2023' column.
bottom_5_by_market_value = df.nsmallest(5, 'Market value as of 31 December, 2023')
print("\nBottom 5 Companies by Market Value:")
# Display the 'Company' and 'Market value as of 31 December, 2023' columns for the bottom 5 companies.
display(bottom_5_by_market_value[['Company', 'Market value as of 31 December, 2023']])

# 6. Calculate the correlation matrix for numerical columns to understand the relationships between these variables.
# Select only numerical columns (int64 and float64) for correlation calculation.
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
# Calculate the pairwise correlation of columns using the selected numerical columns.
correlation_matrix = df[numerical_cols].corr()
print("\nCorrelation Matrix for Numerical Columns:")
# Display the resulting correlation matrix.
display(correlation_matrix)
```

```python
# Import libraries for plotting: matplotlib.pyplot for basic plotting and seaborn for enhanced visualizations.
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Create a Bar plot to visualize the number of companies in each domain.
plt.figure(figsize=(10, 6)) # Set the figure size for better readability.
sns.barplot(x=companies_by_domain.index, y=companies_by_domain.values) # Create the bar plot using domain names on x-axis and counts on y-axis.
plt.title('Number of Companies in Each Domain') # Add a title to the plot.
plt.xlabel('Domain') # Label the x-axis.
plt.ylabel('Number of Companies') # Label the y-axis.
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better visibility if domain names are long.
plt.tight_layout() # Adjust layout to prevent labels overlapping.
plt.show() # Display the plot.

# 3. Create a Bar plot to visualize the average score for companies in each domain.
plt.figure(figsize=(10, 6)) # Set the figure size.
sns.barplot(x=average_score_by_domain.index, y=average_score_by_domain.values) # Create the bar plot using domain names on x-axis and average scores on y-axis.
plt.title('Average Score by Domain') # Add a title.
plt.xlabel('Domain') # Label x-axis.
plt.ylabel('Average Score') # Label y-axis.
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels.
plt.tight_layout() # Adjust layout.
plt.show() # Display the plot.

# 4. Create a Scatter plot to visualize the relationship between 'Market value as of 31 December, 2023' and 'Score'.
plt.figure(figsize=(10, 6)) # Set the figure size.
sns.scatterplot(data=df, x='Market value as of 31 December, 2023', y='Score') # Create the scatter plot.
plt.xscale('log') # Use logarithmic scale for the x-axis ('Market Value') due to its wide range, making the distribution clearer.
plt.title('Market Value vs. Score (Log Scale for Market Value)') # Add a title indicating the log scale.
plt.xlabel('Market value as of 31 December, 2023 (Log Scale)') # Label x-axis, indicating log scale.
plt.ylabel('Score') # Label y-axis.
plt.tight_layout() # Adjust layout.
plt.show() # Display the plot.

# 5. Create a Histogram of the 'Score' distribution to show the frequency of scores within different ranges.
plt.figure(figsize=(10, 6)) # Set the figure size.
sns.histplot(data=df, x='Score', bins=10, kde=True) # Create the histogram with 10 bins and a Kernel Density Estimate line.
plt.title('Distribution of Scores') # Add a title.
plt.xlabel('Score') # Label x-axis.
plt.ylabel('Frequency') # Label y-axis.
plt.tight_layout() # Adjust layout.
plt.show() # Display the plot.

# 6. Create a Heatmap of the correlation matrix to visualize the pairwise correlations between numerical variables.
plt.figure(figsize=(8, 6)) # Set the figure size.
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f") # Create the heatmap, annotating with values, using 'coolwarm' colormap, and formatting values to 2 decimal places.
plt.title('Correlation Matrix of Numerical Variables') # Add a title.
plt.tight_layout() # Adjust layout.
plt.show() # Display the plot.
```
